7. What are the advantages of using generators over regular functions ?
- Advantages of generators over regular functions are >>
  - Memory efficient - one of the biggest advantages of generators is that they are memory efficient since a generator yeilds its values one by one , it does not need to store all of the values in memory at once.
  - Readability - Generators can make your code easier to read and understand . by breaking up the iteration into smaller chunks , you can write code that is more concise and easier to follow.
  - Lazy Evaluation - Generators use lazy evaluations , they don't compute all values upfront , they only compute the next value when requested.
  - Speed - Generators can also be faster than regular functions , especially when working with large data sets. this is because they avoid the overhead of creating a list or other data structure to store the result.
  - Infinite Sequences - Generators are ideal for representing infinte sequences , like a stream of data or a never ending series of numbers , without running out of memories.

```python
#1 Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_of_even_numbers(numbers):
    """
    Calculates the sum of even numbers in a list using a loop.

    Args:
    numbers (list): A list of integers.

    Returns:
    int: Sum of even numbers in the list.
    """
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:  # Check if the number is even
            even_sum += num
    return even_sum


numbers = [10, 15, 20, 25, 10]
result = sum_of_even_numbers(numbers)
print(f"The sum of even numbers is: {result}")

```

```python
#3 Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
def square_numbers(numbers):
    """
    Returns a list containing the squares of each number in the given list.

    Args:
    numbers (list): A list of integers.

    Returns:
    list: A new list with squares of each integer.
    """
    return [num ** 2 for num in numbers]


numbers = [3,4,5,6,7]
result = square_numbers(numbers)
print(f"The squares of the numbers are: {result}")

```

```python
#4 Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(number):
    """
    Checks if a number is prime.

    Args:
    number (int): The number to check.

    Returns:
    bool: True if the number is prime, False otherwise.
    """
    if number <= 1:
        return False
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False
    return True

def check_and_print_primes():
    """
    Takes a number input from the user and checks if it is prime.
    Prints all prime numbers from 1 to 200.
    """
    try:
        num = int(input("Enter a number to check if it is prime (from 1 to 200): "))
        if num > 200 or num < 1:
            print("Please enter a number between 1 and 200.")
        else:
            if is_prime(num):
                print(f"{num} is a prime number.")
            else:
                print(f"{num} is not a prime number.")

        primes = [n for n in range(1, 201) if is_prime(n)]
        print("Prime numbers from 1 to 200:")
        print(primes)

    except ValueError:
        print("Invalid input. Please enter a valid number.")


check_and_print_primes()





```

```python
#5 Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    """
    Iterator class to generate Fibonacci sequence up to a specified number of terms.
    """
    def __init__(self, max_terms):
        """
        Initialize the iterator with the maximum number of terms.

        Args:
        max_terms (int): The number of terms to generate in the sequence.
        """
        self.max_terms = max_terms
        self.current_term = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current_term >= self.max_terms:
            raise StopIteration
        if self.current_term == 0:
            self.current_term += 1
            return 0
        elif self.current_term == 1:
            self.current_term += 1
            return 1
        else:
            fib = self.a + self.b
            self.a, self.b = self.b, fib
            self.current_term += 1
            return fib


num_terms = 15
fibonacci_sequence = FibonacciIterator(num_terms)
print(f"Fibonacci sequence up to {num_terms} terms:")
for num in fibonacci_sequence:
    print(num, end=" ")
```

```python
#6 Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(max_exponent):

    """
    Args:
    max_exponent (int): The maximum exponent to calculate powers of 2.

    Yields:
    int: The next power of 2 in the sequence.
    """
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent


max_exponent = 10
print(f"Powers of 2 up to 2^{max_exponent}:")
for power in powers_of_two(max_exponent):
    print(power, end=" ")
```

```python
#7  Implement a generator function that reads a file line by line and yields each line as a string.
def file_reader(file_path):
    """
    Generator function that reads and yields lines from a file.

    Args:
    file_path (str): The path to the file.

    Yields:
    str: Each line in the file.
    """
    try:
        with open(file_path, 'r') as file:
            while True:
                line = file.readline()
                if not line:
                    break
                yield line.strip()
    except FileNotFoundError:
        yield "Error: File not found."
    except Exception as e:
        yield f"An error occurred: {e}"




file_path = "/sample.txt"
for line in file_reader(file_path):
    print(line)
```

```python
#9 Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temperatures = [0, 10, 25, 30, 70,90, 100]

# Use map to apply the conversion function to each temperature
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Output
print("Temperatures in Fahrenheit:")
print(fahrenheit_temperatures)

```

```python
#10 Create a Python program that uses 'filter()' to remove all the vowels from a given string.
def remove_vowels(s):
    """
    Removes vowels from the given string using filter().

    Args:
    s (str): The input string.

    Returns:
    str: The string without vowels.
    """
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda x: x not in vowels, s))


input_string = "Harry Potter"
result = remove_vowels(input_string)
print(f"String after removing vowels: '{result}'")
```

```python
'''Q11 - Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:

Order Number                     Book title and author                             quant                       Price per item

34587.                           Learning Python , Mark lutz.                         4                            40.95

98762                            Programming Python, Mark lutz.                       5                            56.80

77226                            Head First Python , Paul barry.                      3                            32.95

88112                            Einfurung in Python3 , Bernd klein.                  3                            24.99

Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,- € if the value of the order is smaller than 100,00 €.

Write a Python program using lambda and map.'''


#solution

# Input list of orders
orders = [
    [34587, "Learning Python, Mark Lutz", 4 , 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einfuhrung in Python3, Bernd Klein", 3, 24.99]
]

# Function to compute total price and apply surcharge if needed
process_order = lambda order: (order[0], order[2] * order[3] + 10 if order[2] * order[3] < 100 else order[2] * order[3])

# Use map to apply the function to each order
result = list(map(process_order, orders))

# Output
print("List of order totals:")
for order in result:
    print(order)
```
